```python
from lxml import html 
import requests 
import requests
import urllib.request
import time
from bs4 import BeautifulSoup,SoupStrainer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import re
import time
from datetime import datetime
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
from urllib.request import urlopen
import lxml
import lxml.etree
import httplib2
import urllib
from string import ascii_lowercase
from pandas import DataFrame
```

```python
url_2= 'https://www.crowdai.org/challenges/dark-skies-classification-of-nighttime-images/leaderboards'
response_2 = requests.get(url_2)
content = response_2.content
soup = BeautifulSoup(content,'html.parser')
tree = html.fromstring(content)
```

```python
URL = [] 
for i in range(1,3):
    url_2 = 'https://www.crowdai.org/challenges?challenge_filter=all&page='+str(i)
    response_2 = requests.get(url_2)
    content = response_2.content
    soup = BeautifulSoup(content,'html.parser')
    tree = html.fromstring(content)
    URL_2 = []
    for i in range(1,40,2):
        a = soup.find_all('a',{'href': re.compile(r'/challenges/(.*?)')})[i].attrs['href']
        URL_2.append('https://www.crowdai.org/'+ ''.join(map(str,a)))
    URL.append(URL_2)
URL = URL[0]+URL[1]
df_URL = pd.DataFrame({'Competation_URL':URL})
```

```python
Dummy_2 = []
for i in range(40):
    url = df_URL.Competation_URL[i]
    response_2 = requests.get(url)
    content = response_2.content
    soup = BeautifulSoup(content,'html.parser')
    tree = html.fromstring(content)
    Dummy = []
    Title = soup.find('div', attrs={'class':'heading-w-icon'}).text
    COMPETITION_STATUS = soup.find('div', attrs={'class':'stat-fl'}).text
    Host = soup.find_all('a',{'href': re.compile(r'/organizers/(.*?)')})[1].text
    Description = soup.find('div', attrs={'class':'article'}).text
    url_leaderboard = 'https://www.crowdai.org/'+soup.find('a',{'href': re.compile(r'/leaderboards(.*?)')}).attrs['href']
    submission = []
    for i in soup.find_all('div', attrs={'class':'stat-fr'}):
        submission.append(i.text)
    if Title is not None:
        Dummy.append(Title)
    else:
        Dummy.append('NA')
    if COMPETITION_STATUS is not None:
        Dummy.append(COMPETITION_STATUS)
    else:
        Dummy.append('NA')
    if Host is not None:
        Dummy.append(Host)
    else:
        Dummy.append('NA')
    if Description is not None:
        Dummy.append(Description)
    else:
        Dummy.append('NA')
    if url_leaderboard is not None:
        Dummy.append(url_leaderboard)
    else:
        Dummy.append('NA')
    Dummy.append(submission[0])
    Dummy.append(submission[1])
    Dummy_2.append(Dummy)
    
df_3= DataFrame(Dummy_2,columns=['Title','COMPETITION_STATUS','Host','Description','url_leaderboard','Submission','Participants']) 
```

```python
Dummy_2 = []
for i in range(40):
    url = df_3.url_leaderboard[i]
    response_2 = requests.get(url)
    content = response_2.content
    soup = BeautifulSoup(content,'html.parser')
    tree = html.fromstring(content)
    Dummy = []
    try:
        a = soup.find_all('a',{'href': re.compile(r'/participants(.*?)')})[5].text
        b = soup.find_all('a',{'href': re.compile(r'/participants(.*?)')})[7].text
        c = soup.find_all('a',{'href': re.compile(r'/participants(.*?)')})[9].text
    except:
        Dummy.append('NA')
    else:
        Dummy.append(a)
        Dummy.append(b)
        Dummy.append(c)
    Dummy_2.append(Dummy)

df_4= DataFrame(Dummy_2,columns=['Winner','Second_Winner','Third_Winner']) 
```

```python
Evaluation_Criteria = []
for i in range(40):
    url = df_3.url_leaderboard[i]
    response_2 = requests.get(url)
    content = response_2.content
    soup = BeautifulSoup(content,'html.parser')
    tree = html.fromstring(content)
    Dummy = []
    for i in soup.find_all('th', attrs={'class':'score'}):
        Dummy.append(i.text)
    Evaluation_Criteria.append(','.join(map(str,Dummy)))

df_5= DataFrame(Evaluation_Criteria,columns=['Evaluation_Criteria']) 
```

```python
df_f = pd.concat([df_3,df_4,df_5,df_URL], axis=1, ignore_index=False)
df_f['Title']=df_f['Title'].str.replace('\n'," ")
df_f['COMPETITION_STATUS']=df_f['COMPETITION_STATUS'].str.replace('\n'," ")
df_f['Description']=df_f['Description'].str.replace('\n'," ")
df_f['Submission']=df_f['Submission'].str.replace('\n'," ")
df_f['Participants']=df_f['Participants'].str.replace('\n'," ")
df_f['Platform_name'] = 'https://www.crowdai.org/'
#df_f.drop(['url_leaderboard'],axis=1,inplace=True)
df_f = df_f[['Platform_name', 'Competation_URL', 'Title', 'Host','Description','COMPETITION_STATUS','Winner','Evaluation_Criteria','Second_Winner','Third_Winner','Submission','Participants']]
```
